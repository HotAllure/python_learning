### 4.2.2 Linear Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 4.2.5 Multiple Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#prepare data with some noise
x1 = np.random.rand(100,1)
x1 = x1*4 - 2

x2 = np.random.rand(100,1)
x2 = x2*4 - 2

y = 3*x1 - 2*x2 + 1 #y = 3x^2- 2
```

```python
#visualize data
plt.subplot(1,2,1)
plt.scatter(x1, y, marker='+')
plt.xlabel('x1')
plt.ylabel('y')

plt.subplot(1,2,2)
plt.scatter(x2, y, marker='+')
plt.xlabel('x2')
plt.ylabel('y')
        
plt.tight_layout()
plt.show()
```

```python
from sklearn import linear_model

#concatenate x1 and x2
x1_x2 = np.c_[x1, x2]

#multiple regression
model = linear_model.LinearRegression()
model.fit(x1_x2, y)

#predict
y_ = model.predict(x1_x2)
```

```python
#visualize data
plt.subplot(1,2,1)
plt.scatter(x1, y, marker='+')
plt.scatter(x1, y_, marker='x')
plt.xlabel('x1')
plt.ylabel('y')

plt.subplot(1,2,2)
plt.scatter(x2, y, marker='+')
plt.scatter(x2, y_, marker='x')
plt.xlabel('x2')
plt.ylabel('y')
        
plt.tight_layout()
plt.show()
```

```python
#show coefficient a and intercept b
print('coeff    :', model.coef_)
print('intercept:', model.intercept_)

#evaluate performance of regression
print('R^2:', model.score(x1_x2, y))
```

## 4.3overfitting - caught in a trap of machine learning




Fit the model to $ y = 4x^{3} - 3x^{2} + 2x + 1 $



```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#prepare data with some noise
x = np.random.rand(100,1)
x = x*2 - 1 

y = 4*x**3 - 3*x**2 + 2*x + 1
y += np.random.randn(100,1) * 2 - 1
```

```python
# split data into Train/Test

#Train data
x_train = x[:30]
y_train = y[:30]

#Test data
x_test = x[30:]
y_test = y[30:]
```

```python
#visualize data
plt.subplot(2,2,1)
plt.scatter(x, y, marker='+')
plt.title('all')

plt.subplot(2,2,3)
plt.scatter(x_train, y_train, marker='+')
plt.title('train')

plt.subplot(2,2,4)
plt.scatter(x_test, y_test, marker='+')
plt.title('test')
        
plt.tight_layout()
plt.show()
```

### fit the model to data with 9th order polynomial

```python
from sklearn import linear_model

order = 9

X_TRAIN = x_train

for o in range(2, order+1):
    X_TRAIN = np.c_[x_train**o, X_TRAIN]
```

```python
#multiple regression
model = linear_model.LinearRegression()
model.fit(X_TRAIN, y_train)
```

```python
#show coefficient and intercept
print('coeff    :\n', model.coef_)
print('intercept:', model.intercept_)

#evaluate performance of regression
print('R^2:', model.score(X_TRAIN, y_train))
```

```python
#visualize result
plt.scatter(x_train, y_train, marker='+')
plt.scatter(x_train, model.predict(X_TRAIN), marker='^')
plt.show()
```

### evaluate model with test data

```python
order = 9

X_TEST = x_test

for o in range(2, order+1):
    X_TEST = np.c_[x_test**o, X_TEST]

print('R^2:', model.score(X_TEST, y_test))
```

```python
#visualize result
plt.scatter(x_test, y_test, marker='+')
plt.scatter(x_test, model.predict(X_TEST), marker='^')
plt.show()
```

## 4.4 coping with overfitting - Regression with penality

```python
from sklearn import linear_model

#regression with penality
model = linear_model.Ridge()
#model = linear_model.Lasso()

model.fit(X_TRAIN, y_train)
```

```python
#show coefficient and intercept
print('coeff    :\n', model.coef_)
print('intercept:', model.intercept_)

#evaluate performance of regression
print('R^2:', model.score(X_TRAIN, y_train))
```

```python
#visualize result
plt.scatter(x_train, y_train, marker='+')
plt.scatter(x_train, model.predict(X_TRAIN), marker='^')
plt.show()
```

```python
print('R^2:', model.score(X_TEST, y_test))

#visualize result
plt.scatter(x_test, y_test, marker='+')
plt.scatter(x_test, model.predict(X_TEST), marker='^')
plt.show()
```

## 4.5 various regression models

```python
import math
import numpy as np
import matplotlib.pyplot as plt

#sin wave
x = np.random.randn(1000, 1)
x = x *5 - 10

y = np.array([math.sin(v) for v in x])
y += np.random.randn(1000)
```

```python
from sklearn import linear_model

#linear regression
model = linear_model.LinearRegression()
model.fit(x, y)

#evaluate performance of model
print(model.score(x, y))
```

```python
#visualize result
plt.scatter(x, y, marker='+')
plt.scatter(x, model.predict(x), marker='o')

plt.show()
```

## 4.5.1 regression with SVM

```python
from sklearn import svm

#SVR : support machine regression
model = svm.SVR()
model.fit(x, y)

#evaluate performance of model
print(model.score(x, y))
```

```python
#visualize result
plt.scatter(x, y, marker='+')
plt.scatter(x, model.predict(x), marker='o')

plt.show()
```

## 4.5.2 regression with random forest

```python
from sklearn import ensemble

model = ensemble.RandomForestRegressor()
model.fit(x, y)

#evaluate performance of model
print(model.score(x, y))
```

```python
#visualize result
plt.scatter(x, y, marker='+')
plt.scatter(x, model.predict(x), marker='o')

plt.show()
```

## 4.5.3 regression with k-nearest neighbor

```python
from sklearn import neighbors

model = neighbors.KNeighborsRegressor()
model.fit(x, y)

#evaluate performance of model
print(model.score(x, y))
```

```python
#visualize result
plt.scatter(x, y, marker='+')
plt.scatter(x, model.predict(x), marker='o')

plt.show()
```
